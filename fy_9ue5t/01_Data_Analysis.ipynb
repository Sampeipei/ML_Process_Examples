### Show 2D image

```python
def fig_image(X, y, target_label, show_num=9):
    X_sub, y_sub = X[y==target_label], y[y==target_label]
    rnd_idx = np.random.permutation(len(y_sub))[:show_num]
    X_sub, y_sub = X_sub[rnd_idx], y_sub[rnd_idx]
    
    rows, cols = np.ceil(show_num/3), 3
    plt.figure()
    print("Showing image with label: {}".format(target_label))
    for i, img in enumerate(X_sub):
        plt.subplot(rows, cols, i+1)
        plt.imshow(img, cmap="gray")
        plt.axis('off')
    plt.show()
```

```python
for label in np.unique(y_train):
    fig_image(X_train, y_train, target_label=label)
```
